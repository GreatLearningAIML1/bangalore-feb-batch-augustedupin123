```python
import pandas as pd
from scipy.stats import ttest_1samp, ttest_ind
df = pd.read_csv("Downloads/data.csv")
df1 = pd.read_csv("Downloads/data1.csv")
df2 = pd.read_csv("Downloads/data2.csv")

df.head()
df1.head()
df2.head()
```

```python
from scipy.stats import ttest_ind
import numpy as np
ttest,pval = ttest_ind(df1,df2)
print(ttest, pval)
```

```python
#pval obtained above (0.5074) is greater than 0.05 so we can assume the NULL hypothesis to be correct, i.e, NO significant difference between the monthly outputs.
```

```python
from statsmodels.stats.power import ttest_power
import numpy as np

power = (np.mean(df1) - np.mean(df2))/std(df1)

print(ttest_power(power, nobs=11, alpha=0.05, alternative='two-sided'))

```
