```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("vehicle.csv")
data.head(2)
```

```python
def score_to_numeric(x):
    if x=='van':
        return 3
    if x=='bus':
        return 2
    if x=='car':
        return 1
```

```python
data['class_num'] = data['class'].apply(score_to_numeric)
```
